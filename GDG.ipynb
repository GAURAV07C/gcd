```python
stemmer = PorterStemmer()
stemmed = []
for word in words:
    print(word, " : ", stemmer.stem(word))
    stemmed.append(stemmer.stem(word))
print(stemmed)
```

```python
from nltk.corpus import stopwords
```

```python
def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
text
```

```python
rem_stop = remove_stopwords(text)
rem_stop
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors
features = ['Sky', 'Size', 'Color']
aeroplane_vector = np.array([1, 5, 3])
bird_vector = np.array([1, 3, 4])

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the vectors
ax.quiver(0, 0, 0, aeroplane_vector[0], aeroplane_vector[1], aeroplane_vector[2], color='r', label='Aeroplane', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, bird_vector[0], bird_vector[1], bird_vector[2], color='b', label='Bird', arrow_length_ratio=0.1)

# Annotate the points
ax.text(aeroplane_vector[0], aeroplane_vector[1], aeroplane_vector[2], "Aeroplane", color='r')
ax.text(bird_vector[0], bird_vector[1], bird_vector[2], "Bird", color='b')

# Set labels and title
ax.set_xlabel('Sky')
ax.set_ylabel('Size')
ax.set_zlabel('Color')
ax.set_title('3D Representation of Document Vectors')

# Set limits for better visibility
ax.set_xlim(0, 6)
ax.set_ylim(0, 6)
ax.set_zlim(0, 6)

# Show legend and plot
ax.legend()
plt.grid()
plt.show()
```
